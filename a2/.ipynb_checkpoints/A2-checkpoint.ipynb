## Access rows with loc[] accessor


```python
# 41.
# https://pandas.pydata.org/docs/user_guide/indexing.html#selection-by-label
# Access rows with loc[] accessor (by labels) using list or slice, here using list
# 41. display row values for 0, 5, and 10 index 'labels'
# Note: loc[] interpreted as 'label' of index--*not* integer position!
# Note: here, they are the first, sixth, and 11th rows

mortality_data.loc[[0, 5, 10]]
```

```python
# 42. verify index 'lables' and row values
mortality_data.head(11)
```

```python
# 43. Access rows using *slice* (loc[] accessor includes stop value)
#generic syntax: loc [start:stop:step]
mortality_data.loc[4:6]
```

```python
# 44. Access rows using *slice*, include start, stop and step values)

mortality_data.loc[0:20:5]
```

```python
# 45. Access rows using conditional expression (Year=1917)
mortality_data.loc[mortality_data.Year == 1917]
```

```python
# 56. demo: simple visualization

import matplotlib.pyplot as plt
import numpy as np

# Note:
# dependent variable(s) (DVs): measured or observed to see if affected by independent variable(s)
# independent variable(s) (IVs): changed or controlled by researcher
# Simply, dependent variable "depends" on independent variable!

# Note: standard plotting convention (research/math/science)
# DVs: plotted on y-axis (vertical axis)
# IVs: plotted on x-axis (horizontal axis)

# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 1, 5, 3])

# create plot
plt.plot(x, y)

# add labels and title (optional)
plt.xlabel("x-axis Label (independent variable)")
plt.ylabel("y-axis Label (dependent variable)")
plt.title("Simple Line Plot")

# display plot
plt.show()
```
